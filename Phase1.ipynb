```python
customer_number = input("Enter Customer Number: ")
customer_name = input("Enter Customer Name: ")
units_consumed = float(input("Enter Units Consumed: "))
```

```python
if units_consumed <= 50:
    bill = units_consumed * 0.50
elif units_consumed <= 150:
    bill = (50 * 0.50) + ((units_consumed - 50) * 0.75)
elif units_consumed <= 250:
    bill = (50 * 0.50) + (100 * 0.75) + ((units_consumed - 150) * 1.20)
else:
    bill = (50 * 0.50) + (100 * 0.75) + (100 * 1.20) + ((units_consumed - 250) * 1.50)

total_bill = bill + (bill * 0.20)
```

```python
print("Customer Number:", customer_number)
print("Customer Name:", customer_name)
print("Units Consumed:", units_consumed)
print("Total Bill Amount:", total_bill)

file = open('electricity_bills.txt', 'a')
file.write(customer_number + "," + customer_name + "," + str(total_bill) + "\n")
file.close()
```
